## 1. Bevezetés és elméleti háttér

### 1.1 A chatbotok helye a modern informatikában

A chatbotok – más néven társalgórobotok – olyan szoftverrendszerek, amelyek természetes nyelvi interfészen keresztül képesek emberi felhasználókkal kommunikálni. Az elmúlt évtizedben a természetes nyelvfeldolgozás (Natural Language Processing, NLP) rohamos fejlődésének köszönhetően e rendszerek egyre kifinomultabbá váltak, és számos alkalmazási területen megjelentek: az ügyfélszolgálattól az egészségügyön át az oktatásig.

A chatbotok két fő architektúrális kategóriába sorolhatók. Az **információ-visszakeresés alapú (retrieval-based)** rendszerek egy előre definiált tudásbázisból választják ki a felhasználó kérdésére leginkább illeszkedő választ. Ezzel szemben a **generatív modellek** (például a nagy nyelvi modellek) képesek korábban nem látott válaszok előállítására is. Jelen dolgozatban az előbbi megközelítést alkalmazom, amely ugyan egyszerűbb, de kontrollálhatóbb és megbízhatóbb eredményeket produkál egy zárt tudásterületen belül.

### 1.2 A TF-IDF algoritmus

A **TF-IDF** (Term Frequency – Inverse Document Frequency) egy statisztikai mérőszám, amelyet széles körben alkalmaznak az információ-visszakeresés és a szövegbányászat területén. Célja, hogy számszerűsítse egy szó fontosságát egy dokumentumban, a teljes dokumentumgyűjtemény (korpusz) kontextusában.

A mérőszám két komponensből áll:

- **TF (Term Frequency)** – a kifejezés gyakorisága: megadja, hogy egy adott szó hányszor fordul elő egy dokumentumban. Minél gyakoribb egy szó az adott dokumentumban, annál magasabb a TF-értéke.

$$TF(t, d) = \frac{f_{t,d}}{\sum_{t' \in d} f_{t',d}}$$

- **IDF (Inverse Document Frequency)** – az inverz dokumentumgyakoriság: azon dokumentumok arányának logaritmusa, amelyekben az adott kifejezés előfordul. Azok a szavak kapnak magas IDF-értéket, amelyek ritkán fordulnak elő a korpuszban, és ezáltal nagyobb megkülönböztető erővel bírnak.

$$IDF(t, D) = \log \frac{|D|}{|\{d \in D : t \in d\}|}$$

A végleges **TF-IDF** érték a két komponens szorzata:

$$TF\text{-}IDF(t, d, D) = TF(t, d) \times IDF(t, D)$$

E megközelítés nagy előnye, hogy automatikusan csökkenti a gyakori, kevéssé informatív szavak (pl. névelők, kötőszavak) súlyát, miközben kiemeli a ritka, tartalmasan megkülönböztető kifejezéseket.

### 1.3 A koszinusz-hasonlóság

A **koszinusz-hasonlóság** (cosine similarity) két nem nulla vektor közötti szögkoszinuszként definiált hasonlósági mérőszám. Az NLP kontextusában a dokumentumok TF-IDF vektorainak összehasonlítására szolgál:

$$\text{cosine\_similarity}(\vec{A}, \vec{B}) = \frac{\vec{A} \cdot \vec{B}}{||\vec{A}|| \times ||\vec{B}||}$$

Az eredmény értéke 0 és 1 között mozog: 0 jelentése, hogy a két vektor ortogonális (nincs közös tartalom), míg az 1-es érték tökéletes egyezést jelent. Ez a metrika különösen alkalmas szöveghasonlóság mérésére, mivel független a dokumentumok hosszától, és kizárólag az irányukat (azaz tartalmi hasonlóságukat) veszi figyelembe.

```python
# --- 3. A KÉRDÉSEK ÉS VÁLASZOK SZÉTVÁLASZTÁSA ---
# A tudásbázis kulcsait (kérdéseket) és értékeit (válaszokat) külön listákba mentjük
kerdesek = list(tudasbazis.keys())
valaszok = list(tudasbazis.values())

# --- 4. TF-IDF VEKTORIZÁLÁS ---
# A TfidfVectorizer objektum létrehozása
# - A vektorizáló a kérdéseket numerikus (TF-IDF) vektorokká alakítja
# - Ez lehetővé teszi a matematikai összehasonlítást a szövegek között
vektorizalo = TfidfVectorizer()

# A tudásbázis kérdéseinek vektorizálása (a modell "betanítása")
# A fit_transform() egyszerre illeszti a modellt és transzformálja az adatokat
kerdes_vektorok = vektorizalo.fit_transform(kerdesek)

print(f"✅ TF-IDF mátrix létrehozva. Méret: {kerdes_vektorok.shape}")
print(f"   → {kerdes_vektorok.shape[0]} dokumentum, {kerdes_vektorok.shape[1]} egyedi kifejezés (feature)")
```

```python
# --- 5. A KÉRDÉS-MEGVÁLASZOLÓ FÜGGVÉNY ---
def valasz_keresese(felhasznaloi_kerdes, kueszob=0.1):
    """
    A felhasználó kérdéséhez megkeresi a legrelevánsabb választ a tudásbázisból.

    Paraméterek:
    -----------
    felhasznaloi_kerdes : str
        A felhasználó által feltett kérdés szövege.
    kueszob : float (alapértelmezett: 0.1)
        A minimális hasonlósági pontszám. Ha a legjobb egyezés ez alá esik,
        a rendszer jelzi, hogy nem talált releváns választ.

    Visszatérési érték:
    ------------------
    tuple: (válasz szövege, hasonlósági pontszám, legjobb egyezés kérdése)
    """
    # A felhasználói kérdés vektorizálása a már betanított modell segítségével
    # A transform() függvényt használjuk (nem fit_transform-ot!), mert a modell
    # már illesztve van a tudásbázis kérdéseire
    felhasznalo_vektor = vektorizalo.transform([felhasznaloi_kerdes])

    # Koszinusz-hasonlóság kiszámítása a felhasználói kérdés és
    # az összes tudásbázis-kérdés között
    hasonlosagok = cosine_similarity(felhasznalo_vektor, kerdes_vektorok)

    # A legnagyobb hasonlósági értékkel rendelkező kérdés indexének megkeresése
    legjobb_index = np.argmax(hasonlosagok[0])
    legjobb_pontszam = hasonlosagok[0][legjobb_index]

    # Ellenőrzés: ha a legjobb egyezés is a küszöb alatt van,
    # jelezzük, hogy nem találtunk releváns választ
    if legjobb_pontszam < kueszob:
        return (
            "Sajnálom, a tudásbázisom alapján nem tudok releváns választ adni erre a kérdésre.",
            legjobb_pontszam,
            kerdesek[legjobb_index]
        )

    return (
        valaszok[legjobb_index],
        legjobb_pontszam,
        kerdesek[legjobb_index]
    )

# --- 6. SZÉP FORMÁZÁSÚ KIÍRÓ FÜGGVÉNY ---
def kerdes_megvalaszolasa(felhasznaloi_kerdes):
    """
    Formázott módon jeleníti meg a felhasználó kérdését, a rendszer válaszát,
    valamint a hasonlósági pontszámot és a legjobban illeszkedő kérdést.
    """
    valasz, pontszam, egyezo_kerdes = valasz_keresese(felhasznaloi_kerdes)

    print("=" * 70)
    print(f"📝 Felhasználó kérdése:  {felhasznaloi_kerdes}")
    print("-" * 70)
    print(f"🔍 Legjobb egyezés:      {egyezo_kerdes}")
    print(f"📊 Hasonlósági pontszám: {pontszam:.4f}")
    print("-" * 70)
    print(f"🤖 Chatbot válasza:\n   {valasz}")
    print("=" * 70)
    print()

print("✅ Függvények definiálva.")
```
